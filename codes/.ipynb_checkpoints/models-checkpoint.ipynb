```python
cols = data.columns

for i in range(1,len(cols)):
    print(cols[i])
    print(data[cols[i]].value_counts())
```

**Selecting skewed columns to stratify train and test on**

- ad_type_id
- advertiser_id
- monetization_channel_id
- line_item_type_id

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.3, random_state=42, 
                               stratify=data[['ad_type_id', 'advertiser_id','monetization_channel_id','line_item_type_id']])
```
